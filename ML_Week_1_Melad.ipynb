## **Example 3**

```python
from collections import Counter
import matplotlib.pyplot as plt

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]


histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()], histogram.values(), 10, edgecolor=(0,0,0))

plt.axis([-5,105,0,5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Grade")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")

plt.show()
```

## **Example 6**

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

## **Example 7**

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.axis("equal")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

## **Example 11**


```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [
    [88, 92, 80, 89, 100],
    [78, 85, 70, 83, 90],
    [90, 91, 85, 87, 93],
    [65, 70, 72, 68, 75],
    [95, 98, 96, 92, 99]
]

subjects = ["Math", "Science", "English", "History", "Art"]
students = ["Alice", "Bob", "Charlie", "David", "Emma"]

sns.heatmap(
    data,
    annot=True,
    fmt="d",
    xticklabels=subjects,
    yticklabels=students,
    cmap="YlGnBu"
)

plt.title("Student Exam Scores Heatmap")
plt.show()
```

## **Example 12**

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
colors = ['red', 'gold', 'pink', 'brown']
explode = (0.1, 0, 0, 0)  # "explode" first slice

plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140
)

plt.title('Fruit Market Share')
plt.axis('equal')  # ensures circle shape
plt.show()
```

**Line Plot: Sine & Cosine**

Task: Plot sin(x) and cos(x) on the same figure with labels, title and axis labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200)
plt.figure()
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Sine & Cosine')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

**Q2) Bar Chart of Category Counts**

Task: Given categories [A,B,C,D] and counts [12,7,19,5], make a bar chart with title and axis labels.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
counts = [12, 7, 19, 5]

plt.bar(categories, counts, color='skyblue')

plt.title('Category Counts')
plt.xlabel('Category')
plt.ylabel('Count')

plt.show()
```

**Q3) Histogram of Random Data**

Task: Generate 1000 samples from Normal(0,1). Plot a histogram with 30 bins and clear axis labels.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, color='lightgreen', edgecolor='black')

plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

**Q4) Scatter with a Trend**

Task: Create x from 0 to 10 (80 points) and y = 0.8x + noise. Plot a scatter, add title and labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 80)
y = 0.8 * x + np.random.normal(0, 1, size=x.size)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='dodgerblue', alpha=0.7, edgecolor='black')

plt.title('Scatter Plot with a Linear Trend')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

**Q5) Pie Chart of Proportions**

Task: Draw a pie chart for labels = [North,South,East,West] and shares = [40,25,20,15] with percentages.

```python
labels = ['North', 'South', 'East', 'West']
shares = [40, 25, 20, 15]

plt.figure()
plt.pie(shares, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Regional Share')
plt.axis('equal')
plt.show()
```
